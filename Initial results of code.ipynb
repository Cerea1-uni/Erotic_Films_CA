```python
import requests
from bs4 import BeautifulSoup
import re

def fetch_imdb_script(movie_title_slug):
    """
    movie_title_slug (str): The URL slug that IMSDb uses for the film, 
    e.g. 'the-matrix' for https://www.imsdb.com/scripts/The-Matrix.html
    Returns:str | None: The plain‑text script if found; otherwise None.
    """
 # 1. Build the IMSDb script URL (take the erotic film Anora as example)
    url = f"https://www.imsdb.com/scripts/Anora.html"
    resp = requests.get(url)
    if resp.status_code != 200:
        print(f"Page not found: {url}")
        return None

    soup = BeautifulSoup(resp.text, "html.parser")
    # Parse HTML with BeautifulSoup; script content is usually in <td class="scrtext"> or in a <pre> tag.
    pre = soup.find("td", class_="scrtext") or soup.find("pre")
    if not pre:
        print("Script container not found")
        return None

    # Extract clean text using newline separator.
    text = pre.get_text(separator="\n")
    return text

if __name__ == "__main__":
    slug = "the-matrix"  #need to manually check the slug
    script = fetch_imdb_script(slug)
    if script:
        #  Split into non‑empty lines for preview
        lines = [line.strip() for line in script.splitlines() if line.strip()]
        # Show first 50 lines as a sample
        for l in lines[:50]:
            print(l)
```

```python
import re
def srt_to_txt(input_path, output_path):
   """
   Convert SRT subtitle file to plain text dialogue file
   Parameters:
       input_path: Input SRT file path
       output_path: Output TXT file path
   """
   with open(input_path, 'r', encoding='utf-8') as f:
       srt_content = f.read()
   
   # Regular expression to match all subtitle blocks
   pattern = re.compile(
       r'\d+\n'                          # Sequence number line
       r'\d{2}:\d{2}:\d{2},\d{3}'        # Start time
       r' --> '
       r'\d{2}:\d{2}:\d{2},\d{3}\n'      # End time
       r'(.+?(?=\n\d+\n|\Z))',           # Capture text content
       re.DOTALL                          # Cross-line matching
   )
   
   # Extract and process all dialogue text
   dialogues = []
   for match in pattern.finditer(srt_content):
       text = match.group(1)
       # Merge multi-line text into a single sentence, preserving original end punctuation
       merged_text = ' '.join(line.strip() for line in text.split('\n'))
       dialogues.append(merged_text)
   
   # Write to output file, one dialogue per line
   with open(output_path, 'w', encoding='utf-8') as f:
       f.write('\n'.join(dialogues))
# Usage example
input_file = "/Users/tong/Desktop/All parts.srt"
output_file = "/Users/tong/Desktop/Dialogue.txt"
srt_to_txt(input_file, output_file)
print(f"Conversion completed! Saved to {output_file}")
```
